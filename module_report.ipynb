```python
from math import pi
import random
rand = random.random()*10
p = round(pi, 2)
print('원의 반지름: ', rand)
print('원주율 pi: ', p)
print(f'반지름 {rand}인 원의 면적은 {p*(rand)**2}')
```

#### 학습 되돌아 보기 (3)

```python
from math import factorial as ft
import statistics

def median(st):
    return statistics.median(st)
def mean(st):
    return statistics.mean(st)
def variance(st):
    return statistics.variance(st)
def stdev(st):
    return statistics.stdev(st)

if __name__ == '__main__':
    for i in range(1, 17, 5):
        print(f'{i}! = {ft(i)}')
    
    print('\n')
    
    st = [80, 99, 77, 65, 92, 74, 82]
    print(st)
    print('중앙 값: ', median(st))
    print('평균: ', mean(st))
    print('분산: ', variance(st))
    print('표준편차: ', stdev(st))
    
```
