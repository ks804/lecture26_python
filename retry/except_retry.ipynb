```python
# 학습 내용 되돌아보기 (1)
try:
    f = float('30f')
except Exception as e:
    print(type(e))
#답: <class 'ValueError'>
try:
    [1, 2, 3][3]
except Exception as e:
    print(type(e))
#답: <class 'IndexError'>
try:
    {'fb': 11, 'bb': 9, 'vb': 6}['foot']
except Exception as e:
    print(type(e))
#답: <class 'KeyError'>
try:
    pl = 'python' + 3
except Exception as e:
    print(type(e))
#답: <class 'TypeError'>
```

```python
#학습 내용 되돌아보기 (2)
try:
    m = int(input('월(month) 입력 > '))
    if not (1 <= m <= 12):
        raise Exception('1 ~ 12 사이의 정수를 입력하세요')

except Exception as e:
    print('예외 발생 {}'.format(e))
else:
    print('%d월: 잘 입력했습니다.' % m)

# 1)5를 입력 했을 때: 5월: 잘 입력했습니다.
# 2)0을 입력 했을 때: 예외 발생 1 ~ 12 사이의 정수를 입력하세요
```

```python
# 학습 내용 되돌아보기 (3)
def divide(x, y) :
    try :
        if y == 0 :
            raise Exception('0으로는 나눌 수 없습니다')
        result = x / y
    except Exception as e :
        print(f'Exception : {e}')
    else :
        print(f'결과 {result}')

divide(3.2, 2)
divide(5.4, 0)
```

```python
# 학습 내용 되돌아보기 (3)
uyear = {1:'freshman', 2:'sophomore', 3:'junior', 4:'senior'}
try:
    user_input = int(input(f'대학교 몇 학년이지요?'))
    if not 1 <= user_input <= 4:
        raise Exception
except Exception as e:
    print(f'예외 발생 이름: {type(e)}')
    print('에외 발생 이름: 1 ~ 4 정수를 입력하세요')
else:
    print(f'{user_input}학년: {uyear[user_input]}')
finally:
    print('예외 처리가 잘되는군요!')


#입력을 올바르게 받았을 때:
#1학년: freshman
#예외 처리가 잘되는군요!


#예외 입력 받았을 때:
#예외 발생 이름: <class 'Exception'>
#에외 발생 이름: 1 ~ 4 정수를 입력하세요
#예외 처리가 잘되는군요!
```
