### 리스트의 항목에 접근하기: 슬라이싱(slicing)

```python
test_list = ['iron', 'cap', 'hulk', 'hawk', 'thor', 'wid']

print(test_list[::3])
print(test_list[3:])
print(test_list[2:4])
print(test_list[:6:3])
print(test_list[::1])
print(test_list[:])
print(test_list[::])

print(test_list[-1:-5:-3])
print(test_list[1:-1:])
print(test_list[-2::-1])
print(test_list[::-2])
print(test_list[::-1])
```

### 리스트의 항목 재배열하기(reverse,sort,sorted)

```python
a_list = ['iron', 'cap', 'hulk', 'hawk', 'thor', 'wid']
print('원본 데이터', a_list)
a_list.reverse()
print('역순 후 데이터', a_list)

a_list.sort()
print('정렬 후 데이터', a_list)

a_list = ['iron', 'cap', 'hulk', 'hawk', 'thor', 'wid']
b_list = sorted(a_list)
print('원본 데이터', a_list)
print('정렬 후 데이터', b_list)

c_list = sorted(a_list, key=len)
print(c_list)
```

```python
al = ['a', 'b', 'c']
print(sorted(al, reverse=True))

print(al)
al.sort(reverse=True)
print(al)
```

### 리스트 내포

```python
# 1~N의 수로 이루어진 리스트 생성
N = int(input("숫자:"))
num_list = []
for i in range(1,N+1):
    num_list.append(i)
print(num_list)
```

### 리스트의 동적 배열

```python
##### 1~N의 수로 이루어진 리스트 생성
N = int(input("숫자:"))
num_list = [i for i in range(1,N+1)]
print(num_list)

# expression
num_list = [i * 2 for i in range(1,N+1)]
print(num_list)

#if문 사용
N = int(input("숫자:"))
num_list = [i for i in range(1,N+1) if i % 2 == 0]
print(num_list)
```
