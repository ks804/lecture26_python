```python
S = input().strip()
res = set()

for i in range(len(S)):
    for j in range(i, len(S)):
        sub_str = S[i:j+1]
        res.add(sub_str)

print(len(res))
```

```python
import sys
input = sys.stdin.readline

n = int(input())

in_person = set()

for _ in range(n):
    name, action = input().split()
    if action == 'enter':
        in_person.add(name)
    else:
        in_person.remove(name)


for person in sorted(in_person, reverse = True):
    print(person)
```

```python
import sys
input = sys.stdin.readline

N = int(input())
log = set()
for i in range(N):
    name, commute = input().split()
    if commute == 'enter':
        log.add(name)
    elif commute == 'leave':
        log.remove(name)
        
log_list = list(log)
log_list.sort(reverse=True)
for employee in log_list:
    print(employee)
```
