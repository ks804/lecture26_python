```python
try:
    num_list = [2, 6, 23, 65, 11]
    index = int(input('인덱스 입력 : '))
    print(num_list[index])
except (ValueError, IndexError):
    print(f'0부터 {len(num_list )-1} 사이의 정수를 입력하세요.')
except Exception as e:
    print(f'문제가 발생했습니다. : {e} -> {type(e)}')
```

```python
try:
    number = int(input('숫자 입력 : '))
except ValueError:
    print('숫자를 입력하지 앖았습니다.')
else:
    print(number)
finally:
    print('끝났습니다.')
```

```python
while True:
    try:
        number = int(input('양의 정수 입력 : '))
        if number <= 0:
            raise ValueError
    except ValueError:
        print('양의 정수를 입력하세요.')
    else:
        sum = 0
        for i in range(1, number+1):
            sum += i
        print(f'>> 1부터 {number}까지의 합 : {sum}')
        break
```
