```python
filename = './sample.txt'

with open(filename, encoding='utf-8') as f_read:
    content = f_read.read()
print(content)
```

```python
with open(filename) as f_read:
    lines = f_read.readlines()

for line in lines:
    print(line)
```

```python
with open(filename) as f_read:
    while True:
        line = f_read.readline()
        if line: print(line.strip())
        else: break
```

```python
fruits = ['사과', '토마토', '바나나']
outfile = 'fruits.txt'
with open(outfile, mode='w', encoding='utf-8') as f_write:
    f_write.write('\n'.join(fruits))
```

```python
with open(outfile, mode='a', encoding='utf-8') as f_write:
    f_write.write('\n'.join(fruits))
```

```python
with open(filename, mode='rb') as f_read:
    content = f_read.read()
print(content)
print(content.decode('utf-8'))
```

#### 학습 내용 되돌아보기 (3)

```python
with open('cp.txt', 'wt') as f:
    f.write('freshman\n')
    f.write('sophomore\n')
    f.write('junior\n')
    f.write('senior\n')
with open('cp.txt', 'rt') as f:
    for grade, name in enumerate(f, start=1):
        print(f'{grade}학년 {name}')
```
