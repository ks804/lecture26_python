```python
class Circle:
    pi = 3.14
    def __init__(self, radius):     # radius: 반지름
        self.radius = radius

    def area(self):
        return self.radius ** 2 * Circle.pi

c = Circle(2.5)
print(c.area())
```

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    
rec = Rectangle(2.1, 3.4)
print(rec.area())
```

```python
class Account:
    '''
    class 안에서 주석처리 할때 사용
    '''
    def __init__(self):
        self.balance = 0
    def deposit(self, amount):  # 입금
        self.balance += amount
    def withdraw(self, amount): # 출금
        self.balance -= amount
    
a1 = Account()
a1.deposit(100)

a2 = Account()
a2.deposit(50)

print('a1 : ', a1.balance)
print('a2 : ', a2.balance)
```

```python
# 매직 메서드
class Person:
    def __init__(self,name, age):
        self.name = name
        self.age = age
    # __str__ : 문자열로 반환
    def __str__(self):
        return f'{self.name}({self.age}살)'
    # __eq__ : bool 반환
    def __eq__(self, other):
        if self.name == other.name and self.age == other.age:
            return True
        return False

p1 = Person('홍길동', 15)
print(p1)

p2 = Person('홍길동', 15)
print(p2)

if p1 == p2:
    print('같은사람')
else:
    print('다른사람')
```

```python
class Student:
    def __init__(self, univ, name):
        self.name = name
        self.univ = univ
    def info(self):
        print('대학: '+ self.univ + ' 이름: ' + self.name)

st = Student('한국대학교', '이성실')
st.info()
```

```python
class Student:
    
    def __init__(self, name, dept, mid, final):
        self.name = name
        self.dept = dept
        self.mid = mid
        self.final = final
    def __str__(self):
        return f'학과: {self.dept}, 이름: {self.name}, 중간: {self.mid}, 기말: {self.final}'
    def grade(self):
        avg = (self.mid + self.final) /2
        if avg >= 90:
            return '학점: A'
        elif avg >= 80:
            return '학점: B'
        elif avg >= 70:
            return '학점: C'
        elif avg >= 60:
            return '학점: D'
        else:
            return '학점: F'
        
st2 = Student('김경철', '기계학과', 89, 90)
print(st2.__str__())
print(st2.grade())
```

```python
class Car:
    def __init__(self, name, year, color):
        self.name = name
        self.year = year
        self.color = color
    def __str__(self):
        return f'자동차 회사: {self.name}, 년식: {self.year}, 색상: {self.color}'
    def __eq__(self, other):
        return self.name == other.name and self.year == other.year and self.color == other.color
    
mycar = Car('현대', 2020, '검정')
yourcar = Car('기아', 2021, '백색')
print(mycar)
print(yourcar)
print(mycar == yourcar)
```
