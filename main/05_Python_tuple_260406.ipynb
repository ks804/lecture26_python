### 튜플 항목에 접근

```python
tp = (1, 2, 3, 4, 5)
print(tp[0])
print(tp[2])
print(tp[1:4])
for i in tp:
    print(i, end=' ')
```

### 튜플 탐색

```python
f = ('apple', 'banana', 'kiwi')
print('banana' in f)
print('strawberry' in f)

print(f.index('kiwi'))
print(f.count('apple'))
```

### 튜플 관련 정보

```python
s = (10, 20, 30, 40)
print(len(s))
print(max(s))
print(min(s))
print(sum(s))
```

### 새로운 튜플 생성

```python
a = (3, 2, 1)
b = (4, 5)

print(tuple(sorted(a)))
print(a+b)
print(b*2)
```
