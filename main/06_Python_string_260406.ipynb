## f string

```python
string = 'ABC'
print(f"|{string:10}|")
print(f"|{string:>10}|")
print(f"|{string:^10}|")
print(f"|{string:-^10}|")
print(f"|{string:*>10}|")
```

```python
number = 1234
print(f"|{number:10}|")
print(f"|{number:<10}|")
print(f"|{number:-^10}|")
print(f"|{number:*>10}|")
print(f"|{number:0>10}|")
print(f"|{number:010}|")
```

### 문자 함수

```python
a = input()
b = input()
n = ord(a)
n2 = ord(b)
n, n2 = n2, n
c = (n - n2)-1
print(c)
```

### 모스 부호 번역기

```python
str = input('알파벳 문자열 입력 :')
str = str.upper()
m = ['·-', '-···', '-·-·', '-··', '·', '··-·', '--·', '····', '··', '·---', '-·-', '·-··', '--', '-·', '---', '·--·', '--·-', '·-·', '···', '-', '··-', '···-', '·--', '-··-', '-·--', '--··']
print('모스 부호 :', end = ' ')
for i in range(len(str)) :
    print(m[ord(str[i]) - 65], end = ' ')
```
