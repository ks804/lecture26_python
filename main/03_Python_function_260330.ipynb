```python
#num에 해당하는 구구단 출력
gugudan(5)

def gugudan(num):
    print(f'[구구단 {num}단 ]')
    for i in range(1,10):
        print(f' {num}x{i} = {num*i}')
```

```python
number = int(input('구구단 입력'))
gugudan(number)
```

```python
#2단부터 9단까지 출력하기
for i in range(2,10):
    gugudan(i)
    print()
```

```python
#return 값 확인

def add_and_mul(a,b):
    return a+b, a*b

result = add_and_mul(3,4)
print(result)

reusult1, result2 = add_and_mul(3,4)
print(reusult1, result2)
```

```python
#지역 변수, 전역 변수
sum = 10
def test(num):
    global sum
    sum += num
test(5)
print(sum)
```
